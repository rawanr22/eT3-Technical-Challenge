```python
import os
from shutil import copy
import csv
from datetime import datetime
import pandas as pd
```

```python
main_folder = '/content/drive/MyDrive/eT3Internship2023/problem1/dairies'
destination_folder = '/content/drive/MyDrive/eT3Internship2023/problem1/images_dataset'

#creating the destination folder if it doesn't already exist
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

#function to find all images and copy image them to the new folder
def copy_images(source_folder, destination_folder):
    for root, dirs, files in os.walk(source_folder):
        for file in files:
            if file.lower().endswith('.jpg'):
                source_file_path = os.path.join(root, file)
                destination_file_path = os.path.join(destination_folder, file)
                copy(source_file_path, destination_file_path)

#calling the function with the root folder and new folder as parameters
copy_images(main_folder, destination_folder)
```

```python
#printing the list of all images in the dataset to make sure they've all been copied
images_list = os.listdir(destination_folder)

for image in images_list:
    print(image)
```

```python
modified_images = []
for file_name in images_list: #looping through all images
    name_parts = file_name.split('-')
    if len(name_parts) > 1:
        modified_name = '-'.join(name_parts[1:]) #removing the first part of the image name, in this case the prefix is considered the brand name
        if modified_name in modified_images: #checking if the name of the image already exists
            base_name, ext = os.path.splitext(modified_name)
            counter = 1
            while f"{base_name}_{counter}{ext}" in modified_images:
                counter += 1
            modified_name = f"{base_name}_{counter}{ext}" #adds an extra counter to duplicate image names
        modified_images.append(modified_name) #adding the modified names to the new list

        old_name = os.path.join(destination_folder, file_name)
        new_name = os.path.join(destination_folder, modified_name)
        os.rename(old_name, new_name) #renaming all images in the dataset to the new modified names

        print(modified_name)
```

```python
images_list = os.listdir(destination_folder)

image_info_list = []

#iterating through the images list and extracting image information
for file_name in images_list:
    image_path = os.path.join(destination_folder, file_name)
    image_size = os.path.getsize(image_path)  # get size in bytes
    last_modified = os.path.getmtime(image_path)  # get timestamp of last modification
    last_modified_date = datetime.fromtimestamp(last_modified).strftime('%Y-%m-%d %H:%M:%S')

    image_info = {
        'Image_Name': file_name,
        'Size_in_Bytes': image_size,
        'Last_Modified': last_modified_date
    } #creating a dictionary with the info of each image
    image_info_list.append(image_info) #adding each dict to the list of all images' info

csv_file_path = '/content/drive/MyDrive/eT3Internship2023/problem1/image_info.csv'

#writing image information to the CSV file
with open(csv_file_path, mode='w', newline='') as csv_file:
    fieldnames = ['Image_Name', 'Size_in_Bytes', 'Last_Modified'] #column names
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    writer.writeheader()
    for image_info in image_info_list:
        writer.writerow(image_info) #writing each dict in the list as a row
```
